```python
import pandas as pd
```

```python
df = pd.read_csv('Coronavirus_Analysis/company_financial_data_41.csv')
df['yyyy-mm-dd'] = pd.to_datetime(df['yyyy-mm-dd'])
```

```python
dates = pd.DataFrame(columns=['yyyy-mm-dd'])
dates['yyyy-mm-dd'] = pd.date_range(start="2019-12-31",end="2020-03-31")
```

```python
i = dates.merge(df, how='left')
i
```

```python
i.interpolate(method='linear', axis=0).to_csv('company_financial_data_filled_in.csv', index=False)
```
